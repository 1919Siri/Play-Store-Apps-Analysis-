```python
# Step 1: Define a function to extract the minimum version
def extract_min_version(version):
    if pd.isnull(version) or version == "Varies with device":
        return np.nan
    else:
        # Extract the version part before "and up" or " - "
        version = version.split(" and up")[0].split(" - ")[0]
        
               # Split the version by '.' and filter out any non-numeric parts
        version_parts = [part for part in version.split('.') if part.isdigit()]
         
        # Create a tuple of integers from the numeric part
        version_tuple = tuple(map(int,version_parts))
        return version_tuple
# Step 2: Apply the function to create a new column for the minimum version
df["Min_Android_Ver"] = df["Android_Ver"].apply(extract_min_version)

# Display the updated DataFrame
print(df[["Android_Ver", "Min_Android_Ver"]])
```
